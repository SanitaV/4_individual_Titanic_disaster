```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df = pd.read_csv('/content/train.csv')
```

```python
titanic_df.head()
```

```python
rows, columns = titanic_df.shape
print(f'Number of rows is {rows}, number of columns ir {columns}')
```

```python
missing_values = titanic_df.isnull().sum()
print(missing_values)
```

```python
titanic_df = titanic_df.drop(columns=['Cabin', 'PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket','Fare', 'Embarked', 'Cabin'], axis=1)
titanic_df
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
```

```python
titanic_df.head()
```

```python
titanic_df['Age'].sum()
titanic_df['Age'].min()
titanic_df['Age'].max()
titanic_df['Age'].mean()

titanic_df['Pclass'].min()
titanic_df['Pclass'].max()
```

```python
gender = titanic_df['Sex'].value_counts().reset_index()
print(gender)
```

```python
classes = titanic_df['Pclass'].value_counts().reset_index()
print(classes)
```

```python
Survived = titanic_df['Survived'].value_counts().reset_index()
print(Survived)
```

```python
plt.hist(titanic_df['Age'], color = 'red', edgecolor='white', alpha=0.7)
plt.title ('Histogram of Age distribution in Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(titanic_df['Age'].mean(), linestyle='dashdot', color = 'grey')
plt.axvline(titanic_df['Age'].min(), linestyle='dashdot', color = 'violet')
plt.axvline(titanic_df['Age'].max(), linestyle='dashdot', color = 'blue')
plt.show()
```
